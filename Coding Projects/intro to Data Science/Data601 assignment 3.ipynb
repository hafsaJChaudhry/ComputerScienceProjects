```python
"""Write a Python notebook that generates a file containing the following data:
•Email addresses  X
•Phone numbers  X
•Home Address  X
•Person's name  X
•Year born. Use realistic values. X
•Number of kids. Use realistic values. X
•Categorical variable: rent or own?
User of your notebook should be able to specify how many entities are to be generated.
Generate data in 2 of the following 3 formats: XML, CSV, or JSON. Your choice!"""

#pip install Faker

import numpy as np 
import pandas as pd 
from faker import Faker


def main():
    entities = (int(input("how many entities do you want to create?")))
    count = 0 
    data = []
    while(count != entities):
        fake = Faker()
        name = fake.name()
        email = fake.email()
        address = fake.address()
        cell = fake.phone_number() #international numbers....dk how to change that
        birthRangeList = list(range(1920,2010))
        birthRange = np.random.choice(birthRangeList)
        kidsRangeList = list(range(0,6))
        kidsRange = np.random.choice(kidsRangeList)
        rentList = ["own","rent"]
        rentRange = np.random.choice(rentList, p=[0.65,0.35])
       
        d = {"Name": name,
            "Email": email,
            "Address": address,
            "Cell": fake.phone_number(),
            "Birth Year": birthRange,
            "Number of Kids": kidsRange,
            "Home Owner or Renter": rentRange}
        count +=1 
        data.append(d)
    
    #make into a data frame 
    df = pd.DataFrame(data)
    print(df)
    
    #import data frame into 2 formats 
    df.to_csv("fakeData.csv")
    df.to_json("fakeData.json")
main()
```
