```python
#root node = all_pages
#nodes = random letters
#attributes = author 
#child nodes = content(HTML), source(URL), date(date) 

import os 
from zipfile import ZipFile
from xml.etree import ElementTree
from bs4 import BeautifulSoup

#1a. unzip file 
zip_file = 'xml_containing_html.xml.zip'
unzip = ZipFile(zip_file)

#1b. use a module to parse the XML
#          os.path.abspath shows full path: /Users/hafsachaudhry/Downloads/data/xml_containing_html.xml
xml_file_name = 'xml_containing_html.xml'
full_file = os.path.abspath(xml_file_name)

xml_tree = ElementTree.parse(full_file)
root = xml_tree.getroot()
#print(root)

#2a. extract the HTML from the XML
#create a list that contains all html elements from the child node: content 

#          html_pages.append(html) = [<Element 'content' at 0x7fad1db74e00>, ... ] OR object's memory location
#          html_pages.append(html.text) = gets the actual text per location 
html_pages = []
for html in root.iter('content'):
    html_pages.append(html.text)
    
count = 0 
html_page = 0
for html in html_pages: 
    #2b. use BeautifulSoup to parse the html
    soup = BeautifulSoup(html,'html.parser') 
    
    #3. report the # of links per each page 
    for link in soup.find_all('a'):
        count+=1 
    link = soup.select('a')
    html_page += 1
    print("html page:", html_page)
    print('   ', "number of links for page: " , len(link),"\n", end='')
print("---------------------------------")
print("number of links in total:",count)




```
